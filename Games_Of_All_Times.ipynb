```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = "https://pt.wikipedia.org/wiki/Lista_de_jogos_considerados_os_melhores_de_todos_os_tempos"
html = urlopen(url)
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
title = soup.title
print(title)
```

```python
text = soup.get_text()
print(text)
```

```python
rows = soup.find_all('tr')
print(rows[:10])
```

```python
for row in rows:
    row_td = row.find_all('td')
print(row_td)
type(row_td)
```

```python
import re

list_rows = []
for row in rows:
    cells = row.find_all('td')
    str_cells = str(cells)
    clean = re.compile('<.*?>')
    clean2 = (re.sub(clean, '',str_cells))
    list_rows.append(clean2)
print(clean2)
type(clean2)
```

```python
df = pd.DataFrame(list_rows)
df.head(10)
```

```python
df1 = df[0].str.split(',', expand=True)
df1.head(10)
```

```python
df1[0] = df1[0].str.strip('[')
df1.head(10)
```

```python
df1[4] = df1[4].str.strip(']')
df1.head(10)
```

```python
col_labels = soup.find_all('th')
```

```python
all_header = []
col_str = str(col_labels)
cleantext2 = BeautifulSoup(col_str, "html.parser").get_text()
all_header.append(cleantext2)
print(all_header)
```

```python
df2 = pd.DataFrame(all_header)
df2.head()
```

```python
df3 = df2[0].str.split(',', expand=True)
df3.head()
```

```python
df3 = df3[[0,1,2,3,4]]
```

```python
df3
```

```python
df3[0] = df3[0].map(lambda x: str(x)[:-1])
df3[0] = df3[0].map(lambda x: str(x)[1:])
df3[1] = df3[1].map(lambda x: str(x)[:-1])
df3[2] = df3[2].map(lambda x: str(x)[:-1])
df3[3] = df3[3].map(lambda x: str(x)[:-1])
df3[4] = df3[4].map(lambda x: str(x)[:-9])
```

```python
df1[0] = df1[0].map(lambda x: str(x)[:-1])
df1[1] = df1[1].map(lambda x: str(x)[:-1])
df1[2] = df1[2].map(lambda x: str(x)[:-1])
df1[3] = df1[3].map(lambda x: str(x)[:-1])
df1[4] = df1[4].map(lambda x: str(x)[:-9])
df1
```
